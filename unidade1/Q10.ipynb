```python
def calcular_minimo_cedulas_formatado(valor):
    cedulas = [100, 50, 20, 10, 5, 2, 1]
    resultado = []

    for cedula in cedulas:
        qtd = valor // cedula
        valor -= qtd * cedula
        resultado.append(f"{qtd} de {cedula}")

    return ", ".join(resultado)

# Exemplo de uso:
valor = 378
saida = calcular_minimo_cedulas_formatado(valor)
print(saida)
```
